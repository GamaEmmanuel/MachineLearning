#### b. (5 pts) The data include 10 years (2006 - 2015) of records for both active and terminated employees. Status Year field shows the year of data, and Status field shows the employment status – ACTIVE or TERMINATED in the corresponding status year. The company is interested in what proportion of the staff are leaving. Compute: 1) the percent of terminated employees out of all employees for each year; 2) average termination rate over the 10 years?

```python
# Step 1: Create a pivot_table indexing STATUS_YEAR and apply to STATUS column
```

```python
pd.pivot_table(turnover, index=["STATUS_YEAR"], values = ["EmployeeID"], columns= ["STATUS"], aggfunc=np.count_nonzero)
```

```python
# Step 2: Based on the pivot_table, find total number of employees each year
```

```python
total_employees = pd.pivot_table(turnover, index=["STATUS_YEAR"], values = ["EmployeeID"], columns= ["STATUS"], aggfunc=np.count_nonzero)
```

```python
total_employees = pd.DataFrame(total_employees.EmployeeID)
```

```python
total_employees.rename_axis("year")
```

```python
total_employees["total_employees"] = total_employees.ACTIVE + total_employees.TERMINATED
```

```python
print(total_employees.total_employees)
```

```python
# Step 3: Now compute the percentage
```

```python
total_employees["turnout_rate"] = (total_employees.TERMINATED / total_employees.total_employees)*100
```

```python
total_employees.turnout_rate
```

```python
# Step 4: Invoke a function to compute average on the calculated percentage.
```

```python
print("The total ", total_employees.turnout_rate.mean())
```

#### c.(5 pts) In addition to the proportion of terminated employees, the company wants to know more about different types of termination. Give a stacked bar chart of terminates, where x-axis is status year, y-axis is number of terminated employees, and different colors in a bar show different termination reasons (‘termreason desc’ field in the data). What do you observe in this plot?

```python
# Step 1: Filter dataframe for the relevant status for this question
```

```python
termination_reasons = pd.crosstab(turnover.STATUS_YEAR, [turnover.STATUS, turnover.termreason_desc])
termination_reasons.TERMINATED
```

```python
# Step 2: Similar to part (b) create pivot table on column termreason_desc
```

```python
termination_reasons = termination_reasons.TERMINATED
termination_reasons.rename_axis = "termination reasons"
```

```python
# Step 3: Plot stacked bar chart using pandas plot bar function
```

```python
termination_reasons.plot.bar(stacked=True)
```

What I can see in this plot is that retirement occopies everytime a smaller proportion of the total turnover. On the other hand, layoffs are the most significant reason for turnover over the last two years, which is relevant because before 2014 layoff was not a significant part of the total turnover.
